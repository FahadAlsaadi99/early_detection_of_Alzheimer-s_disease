```python
import glob
Mild = glob.glob('/kaggle/input/alzheimers-dataset-4-class-of-images/Alzheimer_s Dataset/train/MildDemented/*.*')
Moderate = glob.glob('/kaggle/input/alzheimers-dataset-4-class-of-images/Alzheimer_s Dataset/train/ModerateDemented/*.*')
NonDemented = glob.glob('/kaggle/input/alzheimers-dataset-4-class-of-images/Alzheimer_s Dataset/train/NonDemented/*.*')
VeryMild = glob.glob('/kaggle/input/alzheimers-dataset-4-class-of-images/Alzheimer_s Dataset/train/VeryMildDemented/*.*')


data=[]
labels = []

for i in NonDemented:   
    image=tf.keras.preprocessing.image.load_img(i, target_size= (length,width),grayscale=False)
    image=np.array(image)
    data.append(image)
    labels.append(0)

for i in VeryMild:   
    image=tf.keras.preprocessing.image.load_img(i, target_size= (length,width),grayscale=False)
    image=np.array(image)
    data.append(image)
    labels.append(1)

for i in Mild:   
    image=tf.keras.preprocessing.image.load_img(i, target_size= (length,width),grayscale=False)
    image=np.array(image)
    data.append(image)
    labels.append(2)

for i in Moderate:   
    image=tf.keras.preprocessing.image.load_img(i, target_size= (length,width),grayscale=False)
    image=np.array(image)
    data.append(image)
    labels.append(3)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(rescale=1./255,
                            rotation_range=45,
                            width_shift_range=.35,
                            height_shift_range=.35,
                            horizontal_flip=True,zoom_range=(-1,1))

train_data_gen = datagen.flow(
    x=data1,
    y=labels1,
    batch_size=32,
)
```

```python
test_data_gen = datagen.flow(
    x=data_test,
    y=labels_test
)
```

```python
#print(test_data_gen.x.shape) 
```

```python
train_data_gen.x = train_data_gen.x.reshape(208,length*width*3)

test_data_gen.x = test_data_gen.x.reshape(1279,length*width*3)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf  = RandomForestClassifier(n_estimators=100, max_depth=10000)


clf.fit(train_data_gen.x,train_data_gen.y)


clf_pred = clf.predict(test_data_gen.x)

from sklearn import metrics
accuracy = metrics.accuracy_score(test_data_gen.y,clf_pred)
print("Accuracy: {:.2f}".format(accuracy))
```

```python
from sklearn import metrics
accuracy = metrics.accuracy_score(labels_test,clf_pred)
print("Accuracy: {:.2f}".format(accuracy))

from sklearn.metrics import confusion_matrix, classification_report
RF_CM=confusion_matrix(labels_test,clf_pred)
print('Confusion Matrix: \n', RF_CM)
print(classification_report(labels_test,clf_pred))
```
