```python
length = 420
width = 420
```

```python
import glob
Mild = glob.glob('/kaggle/input/alzheimers-dataset-4-class-of-images/Alzheimer_s Dataset/train/MildDemented/*.*')
Moderate = glob.glob('/kaggle/input/alzheimers-dataset-4-class-of-images/Alzheimer_s Dataset/train/ModerateDemented/*.*')
NonDemented = glob.glob('/kaggle/input/alzheimers-dataset-4-class-of-images/Alzheimer_s Dataset/train/NonDemented/*.*')
VeryMild = glob.glob('/kaggle/input/alzheimers-dataset-4-class-of-images/Alzheimer_s Dataset/train/VeryMildDemented/*.*')


data=[]
labels = []

for i in NonDemented:   
    image=tf.keras.preprocessing.image.load_img(i, target_size= (length,width),interpolation='nearest')
    image=np.array(image)
    data.append(image)
    labels.append(0)

for i in VeryMild:   
    image=tf.keras.preprocessing.image.load_img(i, target_size= (length,width),interpolation='nearest')
    image=np.array(image)
    data.append(image)
    labels.append(1)

for i in Mild:   
    image=tf.keras.preprocessing.image.load_img(i, target_size= (length,width),interpolation='nearest')
    image=np.array(image)
    data.append(image)
    labels.append(2)

for i in Moderate:   
    image=tf.keras.preprocessing.image.load_img(i, target_size= (length,width),interpolation='nearest')
    image=np.array(image)
    data.append(image)
    labels.append(3)
```

```python
Moderate_US = Moderate
Mild_US = Mild[:52]
VeryMild_US = VeryMild[:52]
NonDemented_US = NonDemented[:52]
```

```python
from PIL import Image

data1 =[]
labels1 = []

for i in NonDemented_US: 
    image=tf.keras.preprocessing.image.load_img(i,color_mode='rgb',target_size= (length,width),interpolation='nearest')
    image=np.array(image)
    data1.append(image)
    labels1.append(0)
    
for i in VeryMild_US:   
    image=tf.keras.preprocessing.image.load_img(i,color_mode='rgb',target_size= (length,width),interpolation='nearest')
    image=np.array(image)
    data1.append(image)
    labels1.append(1)

for i in Mild_US:   
    image=tf.keras.preprocessing.image.load_img(i,color_mode='rgb',target_size= (length,width),interpolation='nearest')
    image=np.array(image)
    data1.append(image)
    labels1.append(2)

for i in Moderate_US:   
    image=tf.keras.preprocessing.image.load_img(i,color_mode='rgb',target_size= (length,width),interpolation='nearest')
    image=np.array(image)
    data1.append(image)
    labels1.append(3)
```

```python
import glob
Mild_test = glob.glob('/kaggle/input/alzheimers-dataset-4-class-of-images/Alzheimer_s Dataset/test/MildDemented/*.*')
Moderate_test = glob.glob('/kaggle/input/alzheimers-dataset-4-class-of-images/Alzheimer_s Dataset/test/ModerateDemented/*.*')
NonDemented_test = glob.glob('/kaggle/input/alzheimers-dataset-4-class-of-images/Alzheimer_s Dataset/test/NonDemented/*.*')
VeryMild_test = glob.glob('/kaggle/input/alzheimers-dataset-4-class-of-images/Alzheimer_s Dataset/test/VeryMildDemented/*.*')


data_test = []
labels_test  = []

for i in NonDemented_test:   
    image=tf.keras.preprocessing.image.load_img(i, target_size= (length,width),interpolation='nearest')
    image=np.array(image)
    data_test.append(image)
    labels_test.append(0)

for i in VeryMild_test:   
    image=tf.keras.preprocessing.image.load_img(i, target_size= (length,width),interpolation='nearest')
    image=np.array(image)
    data_test.append(image)
    labels_test.append(1)

for i in Mild_test:   
    image=tf.keras.preprocessing.image.load_img(i, target_size= (length,width),interpolation='nearest')
    image=np.array(image)
    data_test.append(image)
    labels_test.append(2)

for i in Moderate_test:   
    image=tf.keras.preprocessing.image.load_img(i, target_size= (length,width),interpolation='nearest')
    image=np.array(image)
    data_test.append(image)
    labels_test.append(3)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(rescale=1./255,
                            rotation_range=45,
                            width_shift_range=.15,
                            height_shift_range=.15,
                            horizontal_flip=True,zoom_range=0.6)

train_data_gen = datagen.flow(
    x=data1,
    y=labels1,
    #batch_size=32,
)
```

```python
test_data_gen = datagen.flow(
    x=data_test,
    y=labels_test
)
```

```python
print(train_data_gen.x.shape)
print(train_data_gen.y.shape)

print(test_data_gen.y.shape)
```

```python
train_data_gen.x = train_data_gen.x.reshape(208,length*width*3)

test_data_gen.x = test_data_gen.x.reshape(1279,length*width*3)
```

```python
from sklearn.svm import SVC
svc = SVC(kernel='rbf',gamma='scale')
svc.fit(train_data_gen.x, train_data_gen.y)
y2 = svc.predict(test_data_gen.x)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy data is",accuracy_score(test_data_gen.y,y2))
```

```python
from sklearn import metrics
accuracy = metrics.accuracy_score(test_data_gen.y,y2)
print("Accuracy: {:.2f}".format(accuracy))

from sklearn.metrics import confusion_matrix, classification_report
RF_CM=confusion_matrix(test_data_gen.y,y2)
print('Confusion Matrix: \n', RF_CM)
print(classification_report(test_data_gen.y,y2))
```
